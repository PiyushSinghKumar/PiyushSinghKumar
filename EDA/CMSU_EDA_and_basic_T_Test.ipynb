```python
shingles.describe()
```

```python
shingles.info()
```

```python
#T_test one sample for A
t_statisticA, p_valueA = stats.ttest_1samp(shingles['A'], 0.35,nan_policy='omit')
p_valueA=p_valueA/2
print('One sample t test \nt statistic: {0} p value for A: {1} '.format(t_statisticA, p_valueA))
```

```python
#T_test one sample for B
t_statisticB, p_valueB = stats.ttest_1samp(shingles['B'], 0.35,nan_policy='omit')
p_valueB=p_valueB/2
print('One sample t test \nt statistic: {0} p value for B: {1} '.format(t_statisticB, p_valueB))
```

```python
#2 Sample T Test
t_statistic, p_value = stats.ttest_ind(shingles['A'],shingles['B'], nan_policy='omit')
print('2 Sample T test  \nt statistic: {0} p value: {1} '.format(t_statistic, p_value))
```
